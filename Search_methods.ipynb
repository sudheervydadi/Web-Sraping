```python
import requests 
from bs4 import BeautifulSoup 
import csv 
  
URL = "https://www.packtpub.com/all-products/all-books"
r = requests.get(URL) 
  
soup = BeautifulSoup(r.content,'html.parser')
```

```python
soup.find_all("div",attrs={'class':'block block-search algolia-search-block'})
```

```python
print(soup)
```

```python
soup.find("a",attrs={'class':'action next'})
```

```python
len(soup.find_all("li",attrs={'class':'item product product-item'}))
```

```python
soup.find_all("li",attrs={'class':'item product product-item'})[0]
```

```python
all_books_table = soup.find("table",class_="views-view-grid")
```

```python
print(all_books_table)
```

```python
for book_structure in soup.find_all("li",attrs={'class':'item product product-item'}):
    print(book_structure.find('span',attrs={'class':'price-wrapper'}))
    
```

```python
for book_structure in soup.find_all("li",attrs={'class':'item product product-item'}):
    print(book_structure.find('span',attrs={'class':'price-wrapper'}).string)
```

```python
for book_structure in soup.find_all("li",attrs={'class':'item product product-item'}):
    print(book_structure.find('a',attrs={'class':'product-item-link'}).string)
```

find(): This function is used to find the first occurrence

find('p'): This function is used to find the first occurrence of a tag

find(text="newtext"): This function is used to find the first occurrence of text

find(attrs={'id':'value'}): This function is used to find the first occurrence of attributes based on the attribute value

find(class_='value'): This function is used to find the first occurrence with the CSS class value

find_all(): This function is used to find all occurrences based on filter conditions

find_parent(): This function is used to find the first occurrence in parent tags

find_parents(): This function is used to find all occurrences in parent tags

find_sibling(): This function is used to find the first occurrence in sibling tags

find_siblings(): This function is used to find all occurrences in sibling tags

find_next(): This function is used to find the first occurrence of a tag parsed immediately after the current tag

find_all_next(): This function is used to find occurrences of all tags parsed immediately after the current tag

find_previous(): This function is used to find the first occurrence of a tag parsed immediately before the current tag

find_all_previous(): This function is used to find occurrences of all tags parsed immediately before the current tag
